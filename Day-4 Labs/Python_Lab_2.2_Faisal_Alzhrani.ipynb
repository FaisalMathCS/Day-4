### Task 2
- Use "Random" library to do the following:
    1. Create a list of random numbers using list comprehension.
    2. Create a list of squares of the even numbers from random numbers using list comprehension.
    3. Create a list of tuples (number, square) for random numbers using list comprehension.

```python
# write your code here ^_^ 
import random 
#1
rand_nums = [random.randint(1, 20) for _ in range(10)]
#2
square_even = [x**2 for x in rand_nums if x % 2 == 0]
#3
square_rand = [(x, x**2) for x in rand_nums] 

print("Random numbers using List Comperhension: ")
print(rand_nums)


print("Squared Even numbers from the random list using LIst Comperhension: ")
print(square_even)


print("a list of tuples of numbers and theire square from the random numbers, using list Comperhension: ")
print(square_rand)
```

- Create a dictionary from two lists (one with keys and one with values) using dictionary comprehension.



```python
# write your code here ^_^
#1
nums_squares_cubed = {x: (x**2, x**3) for x in range(1, 6)}
print("Using dict Comperhension: ")
print(nums_squares_cubed) 
#2
print("The two lists: ")
names = ["Faisal", "Nawaf", "Fahad"]
hobbies = ["Sports", "Video Games", "Movies"]
print(names)
print(hobbies)
names_hobbies = {name: hobby for name, hobby in zip(names, hobbies)}
print("using Zip and dict Comperhension you can do this")
print(names_hobbies)
```

### Task 4
- Given a dictionary where keys are student names and values are lists of their scores, write a function that updates the scores of students who have achieved a perfect score (100) by adding a bonus of 5 points to each score. Use dictionary comprehension to achieve this.

    - Implement the function add_bonus_to_perfect_scores(student_scores) that adds a bonus of 5 points to each score of students who have achieved a perfect score (100).

```python
# write your code here ^_^

def add_bonus_to_perfect_scores(student_scores):
    
    student_scores = {key: [score + 5 if score == 100 else score for score in scores] for key, scores in student_scores.items()}
    return student_scores
```
